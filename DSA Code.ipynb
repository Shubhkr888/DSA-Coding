```python
'''
Move Zeroes
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.
'''
num1=[]
num2=[]
def num(*args):
    for i in args:
        if i != 0:
            num1.append(i)
    for i in args:
        if i == 0:
            num2.append(i)
    return num1+num2
```

```python
num(0,1,0,3,12)
```
